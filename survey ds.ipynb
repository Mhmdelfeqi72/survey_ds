```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Top 5 most common programming languages
top_langs = df['Q8'].value_counts().nlargest(5).index

# Filter dataset for top languages
filtered_df = df[df['Q8'].isin(top_langs)]

# Group and count participants per education level and language
grouped = filtered_df.groupby(['Q8', 'Q5']).size().reset_index(name='Count')

# Pivot to shape suitable for grouped bar plot
pivot_df = grouped.pivot(index='Q8', columns='Q5', values='Count').fillna(0)

# Plot
pivot_df.plot(kind='bar', figsize=(12, 7))

plt.title('Top Preferred Programming Languages by Education Level', fontsize=16)
plt.xlabel('Programming Language')
plt.ylabel('Number of Participants')
plt.legend(title='Education Level', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
df_cleaned = df[['Q8', 'Q1']].dropna()

def age_to_number(age_str):
    if isinstance(age_str, str) and '-' in age_str:
        start, end = age_str.split('-')
        return (int(start) + int(end)) / 2
    elif isinstance(age_str, str) and '+' in age_str:
        return int(age_str.replace('+', ''))
    else:
        try:
            return float(age_str)
        except:
            return None

df_cleaned['Age'] = df_cleaned['Q1'].apply(age_to_number)
df_cleaned = df_cleaned.dropna()

avg_age_per_lang = df_cleaned.groupby('Q8')['Age'].mean().sort_values(ascending=False)

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
avg_age_per_lang.plot(kind='bar', color='skyblue')

plt.title('Average Age by Preferred Programming Language', fontsize=16)
plt.xlabel('Programming Language')
plt.ylabel('Average Age')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# نفترض إن df_exp يحتوي على الأعمدة: Q6 (الخبرة) و Q8 (اللغات)

# الخطوة 1: دالة تحويل الخبرة إلى أرقام
import re
def exp_to_float(exp):
    if isinstance(exp, str):
        exp = exp.strip()
        if exp == 'I have never written code':
            return 0
        elif '<' in exp:
            return 0.5
        elif '>' in exp:
            return 20
        elif '-' in exp:
            numbers = re.findall(r'\d+', exp)
            if len(numbers) == 2:
                return (float(numbers[0]) + float(numbers[1])) / 2
        else:
            numbers = re.findall(r'\d+', exp)
            if numbers:
                return float(numbers[0])
    return None

# الخطوة 2: تحويل الخبرة
df_exp['Experience'] = df_exp['Q6'].apply(exp_to_float)

# الخطوة 3: تفكيك اللغات (Q8 ممكن تحتوي على أكثر من لغة مفصولة بفاصلة)
df_exploded = df_exp[['Experience', 'Q8']].dropna()
df_exploded = df_exploded.assign(Language=df_exploded['Q8'].str.split(';')).explode('Language')
df_exploded['Language'] = df_exploded['Language'].str.strip()

# الخطوة 4: حساب المتوسط
avg_exp_per_language = df_exploded.groupby('Language')['Experience'].mean().sort_values(ascending=False).head(10)

# الخطوة 5: الرسم البياني
plt.figure(figsize=(10,6))
avg_exp_per_language.plot(kind='bar', color='skyblue')
plt.title('Average Years of Experience by Programming Language')
plt.xlabel('Programming Language')
plt.ylabel('Average Experience (Years)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
